Faça o gráfico de dispersão entre o crescimento do PIB e a inflação anual de 1961 a 2018. Calcule e mostre na tela a correlação entre as duas grandezas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds = pd.read_csv('https://raw.githubusercontent.com/joaozz21/BasesComputacionais/refs/heads/master/dados/inflaAnual.csv')
dsPib = pd.read_csv('https://raw.githubusercontent.com/joaozz21/BasesComputacionais/refs/heads/master/dados/pibAnual.csv')

dsInfla = ds[ds['Ano'] >= 1961]

y = np.array(dsInfla['Inflação'])
x = np.array(dsPib['Variação anual do PIB real (%)'])
m,b = np.polyfit(x,y, deg= 1)

correlacao = dsInfla['Inflação'].corr(dsPib['Variação anual do PIB real (%)'])**2

print(f'Correlação entre Inflação e variação do PIB é de: {round(correlacao*100,1)}%\n')

plt.figure()
plt.scatter(x,y)
plt.plot(x,x*m+b,color='red',label='Regressão Linear')
plt.legend()
plt.title('Relação entre inflação e variação do PIB')
plt.xlabel('Variação do PIB (%)')
plt.ylabel('Inflação (%)')
plt.show()
```
